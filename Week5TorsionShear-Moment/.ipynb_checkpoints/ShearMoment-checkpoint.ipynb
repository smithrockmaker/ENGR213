### Shear Moment Diagrams Library

[Alfredo Carella](https://www.oslomet.no/en/about/employee/alfcar/) in the Mechancial Engineering program has written a simple open source [Shear and Moment diagram generator](https://jose.theoj.org/papers/10.21105/jose.00065) for python. Here is the link to the [pypi project page](https://pypi.org/project/beambending/) and the [beambending documentation](https://alfredocarella.github.io/simplebendingpractice/). I will provide no more than this for the time being but it seems fairly selfexplanatory.

### Installation:

The beambending library is NOT an anaconda package so you will need to install the library using pip. First go to the Environments tab and see if ```pip``` is installed for Anaconda. 

Assuming ```pip``` appears to be installed then open a terminal window in your Jupyterlab tab and type ```conda list``` at the prompt and hit return. A long list of libraries will scroll past. Scroll back up and verify that ```pip``` is indeed on the list. 

Then type ```pip install beambending``` at your terminal prompt to install the library. If the install is successful then you should find beambending in the list when you repeat ```conda list```. If this all goes well you are ready to use the library.

````python
import os
from beambending import Beam, DistributedLoadV, PointLoadH, PointLoadV, x
````

````python
def beam_1():
    """Run beam 1"""
    beam = Beam(9)  # Initialize a Beam object of length 9 m
    beam.pinned_support = 2    # x-coordinate of the pinned support
    beam.rolling_support = 7  # x-coordinate of the rolling support
    beam.add_loads((
                    PointLoadH(10, 3),  # 10 kN pointing right, at x=3 m
                    PointLoadV(-25, 3),  # 20 kN downwards, at x=3 m
                    DistributedLoadV(-12, (4, 6)),  # 10 kN/m, downwards, for 3 m <= x <= 9 m
                    DistributedLoadV(-20 + x**2, (0, 2)),  # variable load, for 0 <= x <= 2 m
                ))
    fig = beam.plot()
````

````python
beam_1()
````
